Scraping wikipedia page and writing it to a dataframe

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',
                          flavor='bs4')
df = df[0]
df
```

```python
df.dtypes
```

```python
df.columns = df.iloc[0]
df = df.reindex(df.index.drop(0))
df.head()
```

```python
#removing Not assigned values which are present in both Borough column and Neighbourhood coloumn 
for i in df.index:
    if str(df['Borough'][i]) == 'Not assigned' and str(df['Neighbourhood'][i]) =='Not assigned':
        df = df.reindex(df.index.drop(i))
df
```

```python
df = df.reset_index(drop=True)
df.head()
```

```python
df['Postcode'] = df['Postcode'].astype(str)
df['Borough'] = df['Borough'].astype(str)
df['Neighbourhood'] = df['Neighbourhood'].astype(str)
```

```python
#putting Borough into Neighbourhood where it is not assigned 
for i in df.index:
    if df['Neighbourhood'][i] == "Not Assigned":
        df['Neighbourhood'][i] = df['Borough'][i]
```

```python
df
```

Putting all neighbourhoods in one row for same postal code


```python
df = df.sort_values('Postcode')
df = df.reset_index(drop = True)
df
for i in df.index:
    
    for j in df.index:
        if df.index[i] != df.index[j] and df['Postcode'][i] == df['Postcode'][j]:
            df['Neighbourhood'][i] = df['Neighbourhood'][i] +', ' + df['Neighbourhood'][j]
            


df
```

```python
a = []
for i in np.arange(0,210) :
    if df['Postcode'][i] == df['Postcode'][i+1] and  df['Postcode'][i] != df['Postcode'][i+2] or  df['Postcode'][i] == df['Postcode'][i+2] :
        a.append(i)
a
```

```python
for i in a:
    df = df.reindex(df.index.drop(i))

df
```

**CLUSTERING**

```python
#CLUSTERING
# set number of clusters
from sklearn.cluster import KMeans

kclusters = 5

df_clustering = df.drop(['Postcode', 'Neighbourhood', 'Borough'], axis= 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
df['Cluster_labels'] = kmeans.labels_
df
```

```python
address = 'Toronto, Ontario, Canada'

geolocator = Nominatim()
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

Plotting

```python
import folium
import matplotlib.cm as cm
import matplotlib.colors as colors
# create map

map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df['Latitude'], df['Longitude'], df['Neighbourhood'], df['Cluster_labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
